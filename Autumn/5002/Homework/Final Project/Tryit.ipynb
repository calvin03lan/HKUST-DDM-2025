# 检查胜利者

```python
def check_winner(board):
    # Input: board = current configuration of the 15x15 u-v matrix
    # Output: 1 or -1 to indicate which color wins; 2 if it is a draw; 0 if the game is not yet over
    size = 15
    # all start points（u+v ≤ 14）
    for u in range(size):
        for v in range(size - u):
            player = board[u][v]
            if player == 0 or player == 5:
                continue
            # horizontal check ([u,v+1]~[u,v+4])
            if v + 4 < size and all(board[u][v + k] == player for k in range(5)):
                return 1 if player == 1 else -1
            # vertical check ([u+1,v]~[u+4,v])
            if u + 4 < size and all(board[u + k][v] == player for k in range(5)):
                return 1 if player == 1 else -1
            # digonal check（u-1,v+1到u-4,v+4）
            if u >= 4 and v + 4 < size and all(board[u - k][v + k] == player for k in range(5)):
                return 1 if player == 1 else -1
    # draw
    if all(board[u][v] != 0 for u in range(size) for v in range(size - u)):
        return 2
    return 0
```
